```python
#assignment on seaborn on tips
import seaborn as sb
import matplotlib.pyplot as plt

tips=sb.load_dataset('tips')
sb.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("scatter plot")
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.show()
print(tips)
```
