```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data=pd.read_csv("/content/drive/MyDrive/DAP_TEAM1/22A81A0505/temp.txt",sep=' ')
days=data['days']
temp=data['temp']
temp1=list(temp)
a=np.array(temp)
mean=round(a.mean())
t_temp,t_days,n_days,n_temp,mean_days,mean_temp=[],[],[],[],[],[]
for i in range(0,len(days)):
  if(temp[i]>mean):
    t_temp.append(temp[i])
    t_days.append(days[i])
  elif(temp[i]<mean):
    n_days.append(days[i])
    n_temp.append(temp[i])
  else:
    mean_days.append(days[i])
    mean_temp.append(temp[i])
plt.scatter(t_days,t_temp,color='red',label='>threshold')
plt.scatter(n_days,n_temp,color='orange',label='<threshold')
plt.plot(mean_days,mean_temp,color='green',label='mean')
plt.scatter(mean_days,mean_temp,color='green')
plt.plot(days,temp,color='black')
plt.legend(loc='best')
plt.show()
days_counts=[len(t_days),len(n_days),len(mean_days)]
color=['red','orange','green']
label=[f'No.of days greater than threshold ({days_counts[0]})',f'No.of days less than threshold ({days_counts[1]})',f'No.of days with avg temperature ({days_counts[2]})']
plt.pie(days_counts,labels=label,colors=color)
plt.show()
x=[a.min(),a.max()]
y=[temp1.index(x[0]),temp1.index(x[1])]
plt.bar(y,x)
plt.xlabel('day')
plt.ylabel('temperature')
plt.title("Minimum and Maximum temperature")
plt.show()
```
